## **Loading the savedModel and Tokenizer**

```python
def load_objects():
    global tokenizer, model, max_sequence_len
    with open('/content/drive/My Drive/Aqsa Abu Bakar(DL-PojectT2a)/Token_Text_generation.pkl', 'rb') as handle:
        tokenizer = pickle.load(handle)

    with open('/content/drive/My Drive/Aqsa Abu Bakar(DL-PojectT2a)/Text_generation_len.pkl', 'rb') as max_len_file:
        max_sequence_len = pickle.load(max_len_file)

    model = tf.keras.models.load_model('/content/drive/My Drive/Aqsa Abu Bakar(DL-PojectT2a)/Text_prediction.h5')
```

```python
load_objects()
```

## **Genrating Text**

```python
def generate_text(seed_text, next_words):
    try:
        next_words = int(next_words)
        generated_text = seed_text

        for _ in range(next_words):
            token_list = tokenizer.texts_to_sequences([seed_text])[0]
            token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
            predicted = model.predict(token_list, verbose=0)
            predicted_word_index = np.argmax(predicted)
            output_word = ''

            for word, index in tokenizer.word_index.items():
                if index == predicted_word_index:
                    output_word = word
                    break

            seed_text += ' ' + output_word
            generated_text += ' ' + output_word

        return generated_text
    except Exception as e:
        return str(e)

iface = gr.Interface(
    fn=generate_text,
    inputs=["text", "text"],
    outputs="text",
    title="Text Generation",
    description="Generate text using a custom-trained model.",
)

iface.launch()
```
